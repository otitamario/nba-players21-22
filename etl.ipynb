```python
import pandas as pd
url = 'https://www.basketball-reference.com/leagues/NBA_2022_per_game.html'
tables = pd.read_html(url)
```

```python
df = tables[0]
df.head()
```

```python
df .info()
```

```python
df.drop(df[df['Age'] == 'Age'].index, inplace = True)
df.shape
```

```python
cols = df.columns
dict_retype={}
for c in cols:
    if c == 'Rk' or c == 'Player' or c == 'Pos' or c == 'Tm':
        pass
    elif c == 'Age' or c == 'G' or c == 'GS':
        dict_retype[c] = 'int64'
    else:
        dict_retype[c] = 'float64'
        


print(dict_retype)
df = df.astype(dict_retype)
df.info()
```

```python
df.dropna(inplace = True)
df.info()
```

```python
df.drop(columns='Rk', inplace = True)
df.head()
```

```python
filename = 'NbaPlayersStats.csv'
df.to_csv(filename)
```
